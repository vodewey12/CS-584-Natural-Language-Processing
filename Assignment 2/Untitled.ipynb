```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('A2ocean_temp.csv', header=None, names=['year', 'ocean_temp', 'avg_temp'])
df.head()
```

```python
df.plot(x='year', y='ocean_temp', kind='line', marker='o', title='Deviation from Ocean Temperature to Average Temperature Over the Years')
plt.xlabel('Year')
plt.ylabel('Ocean Deviation over Average Temperature')
plt.grid(True)
plt.show()
```
